### in this from a given dataset we will try to predict whether a particular input with given features is a rock or a mine. This prediction is extremely important for submarines as it helps them to avoid being in the range of mines and prevent any mishappening


## Importing Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and preprocessing 

```python
#loading sonar dataset(csv file) to pandas dataFrame
sonar_data = pd.read_csv("Copy of sonar data.csv", header = None)
sonar_data.sample(5)
```

```python
#checking Number of rows and column
sonar_data.shape
sonar_data.info()
```

```python
# Describing the data --> Statistical measure of the data 
sonar_data.describe()
```

```python
#checking the collection of mine data and rock data in the dataset
sonar_data[60].value_counts()
```

```python
# we can clearly see that number of rock data and number of mine data are almost equal therefore the datset is not biased 
sonar_data.groupby(60).mean()
#grouping target column on the basis of mean has provided us with the insight that mine data inputts have generally larger value
```

### Seperating Data and labels

```python
#independent features are on X and dependant features are on Y
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
#Splitting the data into training and testing split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state=2)
print(X.shape, x_train.shape, x_test.shape)
print(Y.shape, y_train.shape, y_test.shape)
```

### Model Training --> Logistic Regression Used 

```python
#storing our selected model in the model variable
model = LogisticRegression()

#training the logistic regression model with training data 
model.fit(x_train, y_train)
```

### Model Evaluation

```python
# accuracy on the training data 
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)

print("Accuracy on training data: ", training_data_accuracy)
```

```python
#accuracy on test data 
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)

print("Accuracy on testing data: ", test_data_accuracy)

```

Our model performed even better on unseen data which generally does not happen

### Making a Predictive System

```python
#now you can give any available rock or mine data and it will predict whether it is a mine or rock 

print("Enter the features for the object (comma-separated values):")
# Take the input, split it by spaces, and strip any excess whitespace from each item
input_data = input().strip().split(',')

# Convert the input into a numpy array of floats
# We ensure all elements are properly converted to float
try:
    input_data_as_numpy = np.asarray([float(i) for i in input_data])
except ValueError as e:
    print(f"Error: Could not convert input to numbers. Please ensure the input is numeric. {e}")
    exit()

# Ensure the array is 2D (shape: 1, n_features)
input_data_reshaped = input_data_as_numpy.reshape(1, -1)

# Predict using the trained model
try:
    prediction = model.predict(input_data_reshaped)

    # Interpret the prediction
    if prediction[0] == 'R':
        print("The object is a Rock.")
    else:
        print("The object is a Mine, STAY ALERT!")
except Exception as e:
    print(f"Error during prediction: {e}")
```
